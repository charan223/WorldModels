```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
with open('evo_vae_28_pop_size_800_length_4_avg_rollout.pkl', 'rb') as f:
    info_vae = pickle.load(f)
with open('Evo_vae_trained_lstm_28_4_800.pkl', 'rb') as f:
    info_train = pickle.load(f)
with open('Evo_vae_untrained_lstm_28_4_800.pkl', 'rb') as f:
    info_untrain = pickle.load(f)
```

```python
generation = info_train[2][3]
print (generation) #number generations stored
```

```python
data1 = info_vae[2][2]
data2 = info_train[2][2]
data3 = info_untrain[2][2]
```

```python
maxes1 = np.array(data1).T[0]
avgs1 = np.array(data1).T[1]
mins1 = np.array(data1).T[2]
maxes2 = np.array(data2).T[0]
avgs2 = np.array(data2).T[1]
mins2= np.array(data2).T[2]
maxes3 = np.array(data3).T[0]
avgs3 = np.array(data3).T[1]
mins3 = np.array(data3).T[2]
```

```python
t = np.arange(0., generation ,1)
```

```python
#print(maxes ,avgs, mins)
```

```python
maxes1_ad = maxes1 * 10/8
maxes2_ad = maxes2 * 10/8
maxes3_ad = maxes3 * 10/8
```

```python
%matplotlib
#plt.clear()
fig, ax = plt.subplots()
#ax.plot(t, s)
#ax.plot(t, -maxes, 'r--', t,-avgs, 'b--', t, -mins, 'g--')

ax.plot(t, -maxes1_ad[:generation], 'ro-', label="VAE only")
#ax.plot(t,-avgs1[:generation], 'r^-')
#ax.plot(t, -mins1[:generation], 'r2-')
ax.plot(t, -maxes2_ad[:generation], 'bo-', label="VAE + trained LSTM")
#ax.plot(t,-avgs2[:generation], 'b^-')
#ax.plot(t, -mins2[:generation], 'b2-')
ax.plot(t, -maxes3_ad[:generation], 'go-', label="VAE + untrained LSTM")
#ax.plot(t,-avgs3[:generation], 'g^-')
#ax.plot(t, -mins3[:generation], 'g2-')

plt.xlabel('Generation', size = 24)
plt.ylabel('Score of average rollout', size = 24)
plt.title('Scores of controller for different inputs', size = 28)

ax.legend(loc="lower right", frameon=False, prop={'size': 20}) #, title="Legend"
#size = 20
#ax.set(xlabel='Generations',size = 20, ylabel='Average rollouts score',
#       title='VAE-Only controller training')
#fig.savefig('plot_vae.png')#, bbox_inches='tight')
#plt.plot(t, -maxes, 'ro', t,-avgs, 'bo', t, -mins, 'go')
```
